```python
'''
we need to materialize in some way the baskets of actors for every movie
first we need a table like : 
| movieKey1 | actorKey1 |
| movieKey1 | actorKey2 |
| movieKey1 | actorKey3 |
....

second we need to group the result using the keys of movies, the desiderata is something like this:
| movieKey1 | actorKey1, actorKey2, actorKey3, ... |
| movieKey2 | actorKey1, actorKey4, actorKey8, ... |

since we have the keys of the person acting ('nconst') in the film I don't think we need the table people for now

we use the letfsemi join
this type of join is keeping all the rows of the left table in which there is 
at least a match in the right table.
so, in our case, we have the table relationActors in which there are stored the
actor / actresses of every title but we wanna keep only the roles played in movies.
so, if we used the leftsemi join with left-table relationActors and right-table movies
we keep only the rows actors/actresses who played in movies.

the query that the leftsemi join builds is something like
SELECT *
FROM relationActors rA
WHERE EXISTS (SELECT * FROM movies M WHERE rA.tconst = M.tconst)
'''
tmpBaskets = relationActors.join(movies,'tconst','leftsemi')
```

```python
'''
first we aggregate the result with groupBy function on the movies key;
  second we aggregate with .agg on only the relevant column: actor / actress key;
    - DataFrame.agg: Aggregate on the entire DataFrame without groups (shorthand for df.groupBy().agg());
    third we use a set based notion to remove duplicates if any;
      - pyspark.sql.functions.collect_set(col): Aggregate function: returns a set of objects with duplicate elements eliminated;

i looked at another implementation on Kaggle for these operations regarding data organization:
  https://www.kaggle.com/surtrsurtr/imdb-dataset-2
'''
from pyspark.sql.functions import collect_set
tmpBaskets = tmpBaskets.groupBy('tconst').agg(collect_set('nconst').alias('items'))
```

```python
from itertools import combinations
import time

'''
retrieve frequent singletons

transform the list of lists of baskets into a single list with every single item 
repeated as many times as the number of times it shows up in every single basket
  map every to entry to tuple (key,1);
  reduce to (key,count);
  check that it is actually frequent.

return a list of frequent singletons and a dictionary with 'key' : count for 
every singleton found in baskets with his support to calculate association rules easily
'''
def frequentSingletons(RDD, threshold):
  print("Processing and counting singleton candidates...")
  listFreqSingletons = (
      RDD
      .flatMap(list)
      .map(lambda a : (a,1))
      .reduceByKey(lambda a1, a2 : a1+a2)
      .filter(lambda a : a[1] >= threshold)
      .map(lambda a : (a[0]))
      )
  singletonsCount = (
      RDD
      .flatMap(list)
      .map(lambda a : (a,1))
      .reduceByKey(lambda a1, a2 : a1 + a2)
      .collect()
      )
  singletonsSupport = dict()
  for i in range(len(singletonsCount)):
    singletonsSupport[singletonsCount[i][0]] = singletonsCount[i][1]

  return listFreqSingletons, singletonsSupport



'''
retrieve combinations of singletons / pairs / triples and so on

kItemSets of dimension 2, k=2 easy case
kItemSets of dimension 3 and more, check for duplication and remove it
'''
def create_kItemSets(frequentItems, k):
  print("Creating candidates of " + str(k) +"-sized item sets...")
  kItemSets = []
  for item1, item2 in combinations(frequentItems[k-1],2):
    tmp = [item1,item2]
    toAdd = []
    if k > 2:
      for l in range(len(tmp)):
        for i in range(len(tmp[l])):
          if tmp[l][i] not in toAdd:
            toAdd.append(tmp[l][i])
    else:
      toAdd = tmp
    if len(toAdd) == k and toAdd not in kItemSets:
      kItemSets.append(toAdd)

  return kItemSets



'''
since I didn't found a better implementation using .map and .reduce with RDD
this function counts occurrences of candidates in baskets

for this function i looked at this implementation:
  http://ethen8181.github.io/machine-learning/association_rule/apriori.html
i couldn't implement or find a better one

return (key,1) tuple for every candidate found in every single basket
'''
def count_kItemSets(basket, listCandidates):
  for i in listCandidates:
    item = set(i)
    if item.issubset(set(basket)):
      return ((tuple(i),1))



'''
count the occurrences of k-sized item sets and found out which one are frequent

return a list of frequent k-item sets and a dictionary with every candidate and its count
'''
def findFrequent_kItemSets(RDD, listCandidates, s):
  print("Processing and counting candidates...")
  listFrequent_kItemSets = (
      RDD
      .map(lambda a : count_kItemSets(a,listCandidates))
      .filter(lambda a : a)
      .reduceByKey(lambda a1, a2 : a1+a2)
      .filter(lambda a : a[1] >= s)
      )
  kItemSetsCount = (
      RDD
      .map(lambda a : count_kItemSets(a,listCandidates))
      .filter(lambda a : a)
      .reduceByKey(lambda a1, a2 : a1+a2)
      .collect()
      )
  kItemSetsSupport = dict()
  for i in range(len(kItemSetsCount)):
    kItemSetsSupport[kItemSetsCount[i][0]] = kItemSetsCount[i][1]

  return listFrequent_kItemSets, kItemSetsSupport


'''
implementation of main function
'''
def A_Priori(RDD, threshold):
  B = RDD
  s = threshold
  k = 1
  frequentItems = dict()
  itemsSupport = dict()

  listFreqSingletons, singletonsSupport = frequentSingletons(B, s)
  listCandidates = listFreqSingletons.collect()
  frequentItems[k] = listCandidates
  itemsSupport[k] = singletonsSupport

  while frequentItems[k]:
    k += 1
    listCandidates = create_kItemSets(frequentItems, k)
    listFrequent_kItemSets, kItemSetsSupport = findFrequent_kItemSets(B, listCandidates, s)

    frequentItems[k] = listFrequent_kItemSets.collect()
    itemsSupport[k] = kItemSetsSupport

  return frequentItems, itemsSupport, k
```

```python
'''
A-priori
  "entirely" from scratch
'''
from itertools import combinations
import time

def singletonsSupport(baskets):
  print("Creating singletons...")
  itemSupport = {}
  for b in range(len(baskets)):
    for i in range(len(baskets[b][0])):
      item = baskets[b][0][i]
      if item not in itemSupport:
        itemSupport[item] = 1
      else:
        itemSupport[item] += 1
  
  return itemSupport

def findFrequent(kSupport, k, s):
  print("Filtering item sets of dimension " + str(k) + "...")
  listFrequent = []
  for i in kSupport[k]:
    if kSupport[k][i] >= s:
      listFrequent.append(i)

  print("Frequent item sets of dimension " + str(k) + ": " + str(len(listFrequent)))
  print(listFrequent)
  return listFrequent

def create_kItemSets(kFrequent, k):
  print("Creating item sets of dimension " + str(k) + "...")
  listCandidates = []
  for item1, item2 in combinations(kFrequent[k-1],2):
    tmp = [item1,item2]
    toAdd = []
    if k > 2:
      for l in range(len(tmp)):
        for i in range(len(tmp[l])):
          if tmp[l][i] not in toAdd:
            toAdd.append(tmp[l][i])
    else:
      toAdd = tmp
    if len(toAdd) == k and toAdd not in listCandidates:
      listCandidates.append(toAdd)

  print("Candidates item sets of dimension " + str(k) + ": " + str(len(listCandidates)))
  return listCandidates

def kItemSetsSupport(baskets, listCandidates, k):
  print("Counting " + str(k) + "-sized item sets...")
  startTime = time.time()
  kItemSetsSupport = {}
  for b in range(len(baskets)):
    setB = set(baskets[b][0])
    for i in listCandidates:
      setI = set(i)
      I = tuple(i)
      if setI.issubset(setB):
        if I not in kItemSetsSupport:
          kItemSetsSupport[I] = 1
        else:
          kItemSetsSupport[I] +=1
          
  print("--- %s seconds ---" % (time.time() - startTime))
  return kItemSetsSupport


def scratch_APriori(baskets, threshold):
  k = 1
  kSupport = dict()
  kFrequent = dict()

  print("Passes on baskets data number " + str(k))
  kSupport[k] = singletonsSupport(baskets)
  kFrequent[k] = findFrequent(kSupport, k, s)
  while kFrequent[k]:
    k += 1
    listCandidates = create_kItemSets(kFrequent, k)
    print("Passes on baskets data number " + str(k))
    kSupport[k] = kItemSetsSupport(baskets, listCandidates, k)
    kFrequent[k] = findFrequent(kSupport, k, s)

  return kFrequent, kSupport, k
```
